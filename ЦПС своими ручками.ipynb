# В задании используется гамма распределение с мат.ожиданием и дисперсией:

```python
#Генерируем выборку
gamma_default = np.random.gamma(k, tetta,1000) 

 # Построение гистограммы
count, x, ignored = plt.hist(gamma_default, 50, density=True, label='Гистограмма выборки')
x = np.linspace(0, 25, 100)
y = x**(k-1)*(np.exp(-x/tetta) /  
(sps.gamma(k)*tetta**k))
plt.plot(x, y, linewidth=2, color='r', label='Теоретическая плотность')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

# Генерация 1000 выборок и построение гистограммы при n=5

```python
#Генерируем выборочное среднее и записываем в means через append
means=[]
for i in range(1000):
    s=np.random.gamma(k, tetta,5).mean() # параметр n=5
    means.append(s)
    
#определяем мат. ожидание mu и дисперсию sigma
mu=k*tetta
dispersia=k*(pow(tetta,2))/5
sigma=np.sqrt(dispersia)

#Построение гистограммы
count, x, ignored = plt.hist(means, 50, density=True, label='Гистограмма построенного нормального распределения для n=5')
x = np.linspace(0, 25, 100)
y = 1/(sigma * np.sqrt(2 * np.pi)) * \
               np.exp( - (x - mu)**2 / (2 * sigma**2) )
plt.plot(x, y, linewidth=2, color='r', label='Теоретическая плотность нормального распределения для n=5')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

# Генерация 1000 выборок и построение гистограммы при n=10

```python
#Генерируем выборочное среднее и записываем в means через append
means=[]
for i in range(1000):
    s=np.random.gamma(k, tetta,10).mean() #параметр n=10
    means.append(s)
    
#определяем мат. ожидание mu и дисперсию sigma    
mu=k*tetta
dispersia=k*(pow(tetta,2))/10
sigma=np.sqrt(dispersia)

#Построение гистограммы
count, x, ignored = plt.hist(means, 50, density=True, label='Гистограмма построенного нормального распределения для n=10')
x = np.linspace(0, 25, 100)
y = 1/(sigma * np.sqrt(2 * np.pi)) * \
               np.exp( - (x - mu)**2 / (2 * sigma**2) )
plt.plot(x, y, linewidth=2, color='r', label='Теоретическая плотность нормального распределения для n=10')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
